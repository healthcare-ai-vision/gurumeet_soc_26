# Image Processing in Python

This notebook demonstrates:

- Grayscale Conversion
- Histogram Visualization
- Image Resizing
- Edge Detection
- Noise Addition
- Noise Removal using Gaussian Filter

Images Used:
1. Animal Image
2. Landscape Image

The images are loaded into Python for further processing.
# animal = cv2.imread('Animal.png')

means:

Find the file named Animal.png...
Read all pixel values.
Store them inside variable animal
After this: animal is no longer a picture. It is actually a huge matrix of numbers.Each group represents pixel color information.


1 row, 2 columns, first position.

OpenCV reads colors as:
Blue Green Red

Matplotlib expects: Red Green Blue

So we convert before displaying.

```python
# OpenCV for image processing
import cv2

# NumPy for numerical operations
import numpy as np

# Matplotlib for displaying images and graphs
import matplotlib.pyplot as plt
```

```python
# Read the animal image
animal = cv2.imread('Animal.png')

# Read the landscape image
landscape = cv2.imread('Landscape.png')
```

```python
# Create a figure window/ drawing area.
plt.figure(figsize=(10,5))

# Display animal image
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(animal, cv2.COLOR_BGR2RGB))
plt.title("Animal")
plt.axis('off')

# Display landscape image
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(landscape, cv2.COLOR_BGR2RGB))
plt.title("Landscape")
plt.axis('off')

# Show both images
plt.show()
```

```python
# Convert animal image to grayscale
animal_gray = cv2.cvtColor(animal, cv2.COLOR_BGR2GRAY)

# Convert landscape image to grayscale
landscape_gray = cv2.cvtColor(landscape, cv2.COLOR_BGR2GRAY)
```

```python
# Print dimensions of original images
print("Animal shape:", animal.shape)
print("Landscape shape:", landscape.shape)

# Print dimensions of grayscale images
print("Animal grayscale shape:", animal_gray.shape)
print("Landscape grayscale shape:", landscape_gray.shape)
```

```python
# Create figure
plt.figure(figsize=(10,5))

# Display grayscale animal image
plt.subplot(1,2,1)
plt.imshow(animal_gray, cmap='gray')
plt.title("Animal Grayscale")
plt.axis('off')

# Display grayscale landscape image
plt.subplot(1,2,2)
plt.imshow(landscape_gray, cmap='gray')
plt.title("Landscape Grayscale")
plt.axis('off')

# Show images
plt.show()
```

```python
# Create figure - histograms
plt.figure(figsize=(12,5))

# Histogram of animal image
plt.subplot(1,2,1)
plt.hist(animal_gray.ravel(), bins=256)
plt.title("Animal Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Number of Pixels")

# ravel() performs conversion from 2D matrix to single long list.

# Histogram of landscape image
plt.subplot(1,2,2)
plt.hist(landscape_gray.ravel(), bins=256)
plt.title("Landscape Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Number of Pixels")

# Show histograms
plt.show()
```

```python
# Image Resizing

# Resize animal image
animal_resized = cv2.resize(animal, (500, 300))

# Resize landscape image
landscape_resized = cv2.resize(landscape, (500, 300))

print("Animal resized shape:", animal_resized.shape)
print("Landscape resized shape:", landscape_resized.shape)
```

```python
# Create figure -- Display Resized Images.
plt.figure(figsize=(10,5))

# Animal image
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(animal_resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Animal")
plt.axis('off')

# Landscape image
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(landscape_resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Landscape")
plt.axis('off')

plt.show()
```

```python
# Edge Detection

# Detect edges in animal image
animal_edges = cv2.Canny(animal_gray, 100, 200)

# Detect edges in landscape image
landscape_edges = cv2.Canny(landscape_gray, 100, 200)
```

```python
# Create figure -- Display Edges
plt.figure(figsize=(10,5))

# Animal edges
plt.subplot(1,2,1)
plt.imshow(animal_edges, cmap='gray')
plt.title("Animal Edges")
plt.axis('off')

# Landscape edges
plt.subplot(1,2,2)
plt.imshow(landscape_edges, cmap='gray')
plt.title("Landscape Edges")
plt.axis('off')

plt.show()
```

```python
# Noise Addition

# Generate Gaussian noise for animal image
animal_noise = np.random.normal(0, 25, animal_gray.shape)

# Generate Gaussian noise for landscape image
landscape_noise = np.random.normal(0, 25, landscape_gray.shape)

# Add noise
animal_noisy = animal_gray + animal_noise
landscape_noisy = landscape_gray + landscape_noise

# Keep values between 0 and 255
animal_noisy = np.clip(animal_noisy, 0, 255)
landscape_noisy = np.clip(landscape_noisy, 0, 255)

# Convert to uint8
animal_noisy = animal_noisy.astype(np.uint8)
landscape_noisy = landscape_noisy.astype(np.uint8)
```

```python
# Display Both Noisy Images

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(animal_noisy, cmap='gray')
plt.title("Noisy Animal")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(landscape_noisy, cmap='gray')
plt.title("Noisy Landscape")
plt.axis('off')

plt.show()
```

```python
# Remove noise using Gaussian filter
animal_denoised = cv2.GaussianBlur(animal_noisy, (5,5), 0)

landscape_denoised = cv2.GaussianBlur(landscape_noisy, (5,5), 0)
```

```python
# Display filtered image

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(animal_denoised, cmap='gray')
plt.title("Animal After Gaussian Filter")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(landscape_denoised, cmap='gray')
plt.title("Landscape After Gaussian Filter")
plt.axis('off')

plt.show()
```
